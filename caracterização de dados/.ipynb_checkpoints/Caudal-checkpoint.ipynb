```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.dates as mdates
import matplotlib.pyplot as plt 
# INOVA > INOVA - ABASTECIMENTO DE ÁGUA > ZMC de BOLHO > Reservatório de Bolho > Caudal BOLHO > Caudal



df = pd.read_excel(r'caudal.xlsx')
df = pd.DataFrame(df, columns=['Tempo', 'Caudal'])
df.dtypes
#df['Tempo'] =  pd.to_datetime(df['Tempo'], format='%d/%m/%Y %H:%M')
#df = df.set_index('Tempo')

df['Caudal'] = pd.to_numeric(df['Caudal'], errors='coerce')
```

```python
year_summary = df.copy()
year_summary['Tempo'] = pd.to_datetime(year_summary['Tempo'])
year_summary.head()
year_summary.dtypes
```

```python
year_summary['year'] = year_summary['Tempo'].dt.year
year_summary['month'] = year_summary['Tempo'].dt.month
year_summary['dayofweek'] = year_summary['Tempo'].dt.dayofweek
```

```python
year_summary.head()
```

```python

year_summary_grouped= year_summary.groupby(['year', 'month']).agg({'Caudal': {'mean', 'std', 'min', 'max', 'median'}})
year_summary_grouped.head(5 * 12)
```
